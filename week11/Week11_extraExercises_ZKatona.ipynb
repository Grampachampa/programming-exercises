# Book quiz 1

A new quiz game on writers and books is under development. For this game, a class called Author is created. An object in the Author class represents a writer and has the following attributes: name, active years and writing language. The class can be seen below.

You can see that the active years attribute is currently stored in the form of a string (e.g., `"1929-1976"`). You will have two tasks:

1. Add a new method to the Author class called `getYears`, which will return the boundaries of the author's active years in a tuple (e.g., `(1929, 1976)`).

2. Add another method to the Author class called `careerLength`, which returns the number of years in which the author was active (e.g., `47` for an author who was active between 1929 and 1976). After adding this method, use it to print the number of active years for the example author given.

Note: you should reuse the `getYears` method inside the `careerLength` method.

Example:

Input:
```python
author = Author("John Steinbeck", "1929-1976", "US English")
```

Output:
<pre>
John Steinbeck was active for 47 years.
</pre>

(Made by: Zsófia Katona)

````python
from typing import Tuple

class Author():
    '''
    Represents an author with the following attributes:
        name: string: the author's name,
        years: string: active years written in the format "xxxx-yyyy",
        language: string: primary working language of the author.
    '''

    def __init__(self, name: str, years: str, language: str) -> None:
        self.name = name
        self.language = language
        self.years = years

## Start adding the new methods AFTER this line.

# Example solution:
    def getYears(self) -> Tuple:
        '''Returns the boundaries of the author's active years in a tuple.'''
        years = self.years.split("-")
        return int(years[0]), int(years[1])

    def careerLength(self) -> int:
        '''Returns the number of years in which the author was active.'''
        length = self.getYears()[1] - self.getYears()[0]
        return length

## Finish adding the new methods BEFORE this line.

author = Author("John Steinbeck", "1929-1976", "US English")

## Print the message here:

# Example solution:
print(f'{author.name} was active for {author.careerLength()} years.')
````

# Book quiz 2

For the same quiz game on books, a class called Book is created. This class has four attributes: title, author, year and original language. The class can be seen below. Your task is the following:

1. Add a method called `contemporary` to the Book class, which takes an object from the class Author as an argument and returns a boolean value: `True` if the book was published when the author was active and `False` otherwise.

2. Use this method to select authors from an example list which are contemporary to the book given, and print this in a nice message.

Note: you should reuse the `getYears` method in the Author class inside the `contemporary` method in the Book class.

Input:
```python
author1 = Author("John Steinbeck", "1929-1976", "English")
author2 = Author("Ernest Hemingway", "1926-1961", "English")
author3 = Author("Virginia Woolf" "1915-1941", "English")

authors = [author1, author2, author3]

book = Book("Ulysses", "James Joyce", 1922, "English")
```

Output:
<pre>
Ulysses is a book contemporary to the following author(s):
Virginia Woolf
</pre>

(Made by: Zsófia Katona)

````python
class Book():
    '''
    Represents a book with the following attributes:
        title: string: title of the book,
        author: string: name of the author of the book,
        year: integer: year the book was written,
        language: string: original language of the book.
    '''

    def __init__(self, title: str, author: str, year: int, language: str) -> None:
        self.title = title
        self.author = author
        self.year = year
        self.language = language

## Start adding the contemporary method AFTER this line.

    # Example solution:
    def contemporary(self, author: Author) -> bool:
        '''Returns True if the book was published while the author was active, False otherwise.'''
        
        if author.getYears()[0] <= self.year <= author.getYears()[1]:
            return True

        return False

## Finish adding the contemporary method BEFORE this line.

author1 = Author("John Steinbeck", "1929-1976", "English")
author2 = Author("Ernest Hemingway", "1926-1961", "English")
author3 = Author("Virginia Woolf", "1915-1941", "English")

authors = [author1, author2, author3]

book = Book("Ulysses", "James Joyce", 1922, "English")

## Choose and print contemporary authors here:

# Example solution:
print(f'{book.title} is a book contemporary to the following author(s):')

for author in authors:
    if book.contemporary(author):
        print(author.name)
````
