## X. Horse Betting

Want to play a betting game but don't want to lose any real money? Create it yourself!
You start with a wallet that contains 100. Each round you get to choose on which horse you want to bet. You also choose the amount you want to bet. 
If the horse you chose is in first place (after shuffling the list) then the amount you put in, will be doubled. Otherwise you lose your money.

The bets are prompted using the input() function. The player will be asked put in a bet continuously without having to rerun the code every time. If the amount in your wallet is 0 or the player has as input -1, the betting will end immediately.

**Hint:**

To shuffle a list, you first have to import the random library with ``import random``. Then you can shuffle a list using ``random.shuffle(horses)``

**Notes:**
1. Create the function ``start_race()``, which takes the horse of choice, amount that has been betted on this horse, and the total amount that is in the wallet as input.
2. First a list of the horses with numbers in front will be printed.
3. It will then ask the player to insert the number of the horse they want to choose. If the input is -1, the bet will stop and an overview (see point 8) will be printed.
    <pre>"You chose: Affirmed."</pre>
4. Then it will ask the player to insert the amount they want to bet:
    <pre>"You have betted: amount."</pre>
    - If the player does not have that amount in their wallet, it will print <pre> "You do not have that amount in your wallet."<pre>
    - Else it will call the function ```start_race()```.
5. Shuffle the list in the function.
6. When the player betted on the correct horese, it will print:
    <pre>"You have won the bet. Horse has finished first. You have won: (amount the player put in * 2). Your wallet now contains: (wallet + (amount the player put in * 2)) "</pre>
7. When the player betted on the wrong horese, it will print:
    <pre>"You have lost the bet.  Horse has finished first. You have lost: (amount the player put in). Your wallet now contains: (wallet - amount the player put in)!</pre>
8. When the player exits, or the wallet contains 0. The following will be printed:
    - If the player has lost money, it will print <pre> Your wallet contains x. You have lost x in total. </pre>
    - If the player has won money, it will print <pre> Your wallet contains x. You have won x in total. </pre>
    - Else if the player has not won but also not lost money, it will print <pre> You did not lose or win anything."</pre>

````python
horses = ["Secretariat", "Alydar", "Citation", "Seabiscuit", "Kelso", "Affirmed"]
````

````python
import random

def start_race(horse_choice : str, amount : int, wallet : int) -> int : 
    """
    Places a bet on a horse. 
    Input: horse_choice : str - the horse the user chose, amount : int - the amount the player has betted on the horse
    Output: None
    If the player guesses correct, their amount betted on the horse will be doubled.
    If not, the player loses all the money.
    """
    random.shuffle(horses)
    if horses[0] == horse_choice:
        wallet = wallet + (amount * 2)
        print("You have won the bet. " + horses[0] + " has finished first. You have doubled your bet and won: " + str(amount * 2) + ". Your wallet now contains: " + str(wallet))
    else:
        wallet -= amount
        print("You have lost the bet.  " + horses[0] + " has finished first. You have lost: " + str(amount) + ". Your wallet now contains: " + str(wallet))
    
    return wallet

amount : int = 0
choice : int = 0
start_wallet : int = 100
wallet : int = start_wallet

while wallet > 0 and amount != -1 and choice != -1:
    i = 0
    for n in horses:
        i += 1
        print( str(i) + ". " + n)

    choice : int = int(input('Please insert the number of the horse you want to bet on'))
    horse_choice = horses[choice - 1] 

    if choice != -1:
        print("You chose: " + horse_choice) 
        amount : int = int(input('Please insert the amount you want to bet'))
        if amount != -1: 
            print("You have betted: " + str(amount) + ".")

            if int(amount) > int(wallet):
                print("You do not have that amount in your wallet.")
            else: 
                wallet = start_race(horse_choice, amount, wallet)

total = wallet - start_wallet
if total > 0:
    print("Your wallet contains " + str(wallet) + ". You have won " + str(total) + " in total.")
elif total < 0:
     print("Your wallet contains " + str(wallet) + ". You have lost " + str(abs(total)) + " in total.")
else: 
    print("Your wallet contains " + str(wallet) + ". You did not lose or win anything.")
````
