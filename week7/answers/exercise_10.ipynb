## Restaurant order

Create a function called take_order which asks the user the following questions...
* How many plates of lasagne would you like?
* How many plates of lobster would you like?
* How many bottles of champaigne would you like?

...and takes the base tip rate as an argument. The function should return a dictionary called order, in which the customer's order is stored. It should also return a float variable called tip_rate, which is the base tip rate plus 5% if the customer ordered champaigne, and the base tip rate otherwise.

Example:

Input:
```python
tip_rate = 0.1
```

<pre>
How many plates of lasagne would you like? 3
How many plates of lobster would you like? 2
How many bottles of champaigne would you like? 1
</pre>

Output:
<pre>
({"lasagne" : 3, "lobster" : 2, "champaigne" : 1}, 0.15)
</pre>

(Made by: Zsófia Katona)

````python
#Example solution:

from typing import Dict, List

def take_order(tip_rate: float) -> (Dict[str, int], float):
    order: dict = {}
    
    lasagnes = int(input("How many plates of lasagne would you like?"))
    lobsters = int(input("How many plates of lobster would you like?"))
    champaignes = int(input("How many bottles of champaigne would you like?"))
    
    order["lasagne"] = lasagnes
    order["lobster"] = lobsters
    order["champaigne"] = champaignes
    
    if champaignes > 0:
        tip_rate += 0.05
    
    return order, tip_rate

print(take_order(0.10))
````

## Check please!

Create a function called check, which takes three parameters,
* the base tip rate,
* a callable function to take a customer's order and adjust the tip rate accordingly,
* a dictionary which stores the prices of items on the menu,

and returns the total price of the order, including tip. Use this output to print a message with the total price.

Note: pass your take_order function, which you defined in the previous task, as the callable argument to the check function.

Example:

Input:
```python
base_tip = 0.1
order = take_order
prices = {"lasagne" : 25, "lobster" : 80, "champaigne" : 100}
```

<pre>
How many plates of lasagne would you like? 1
How many plates of lobster would you like? 1
How many bottles of champaigne would you like? 1
</pre>

Output:
<pre>
Your total is €235.75.
</pre>


(Made by: Zsófia Katona)

````python
# Example solution:

def check(base_tip: float, order: callable, prices: Dict[str, int]) -> float:
    order, tip_rate = order(base_tip)
    
    total: float = 0.0
    
    for item, number in order.items():
        price = prices[item]
        total += price * number
        
    total = total * (1 + tip_rate)
    
    return total

base_tip = 0.1
prices = {"lasagne" : 25, "lobster" : 80, "champaigne" : 100}

total = check(base_tip, take_order, prices)

print("\n" + "Your total is €{price:.2f}.".format(price = total))
````
