## Social bot by an underpaid developer

Oh, its you again, the underpaid employee that made the abomination of *spam-or-ham* classifier; seems like you weren't fired after that Σ(°ロ°). Now, as you are still of use to the company, the boss tells you to write a bot that can **"recognize and remember to make personalized interactions, while also having some human-like elements"**. Since you are still being \*cough\* legally \*cough\* paid minimum wage (seriously, why don't you just quit), you decided to make something that is considered *good enough*. This bot should:
- Ask for name of the current user.
- Ask if the current user wants to add more facts about themselves, or if they would like to hear facts that the robot remembers.
- The bot should store facts about people inside of a dictionary, with keys being the name, and values being lists of facts (as strings).
- The human-like element would be that the bot is forgetful: after every interactions, there is a 0.05 chance that the bot would either forget, or mistake a fact of someone with that of another person.

Description and optional steps:
* Create a class Bot
* In the **\__init\__**, the bot should initiate a dictionary as its memory
* Create a method called **update_mem()** that takes a string as an arguement, and update the dictionary if prompted.
* Create a method called **inquire_name()** that does not take anything as arguments, but returns a string.
* Create a method called **get_fact()** that does not take anything as arguments, but returns a string.
* Create a method called **forget()** that does not take anything as arguments, but change the dictionary's content in someway (human-like element), this method should be called after each update_mem().
* Create a method called **step()** that would prompt the bot to go to the next step in the interaction.

**Notes:**
You can use the module **random** to generete pseudo-random numbers/probabilities/choices.

Please also handle all exceptions that you can think of (value error, e.t.c); you don't have to consider all of them, just think of this as practicing for future projects.

```python
#optional definitions
import random
#// BEGIN_TODO
class Bot:
    def __init__(self):
        self.memory = {}
        self.current_user = self.inquire_name()
        self.options = ["Add a fact", "Facts about myself", "Quit"]
        
    def update_mem(self, fact: str) -> None:
        if self.current_user == None:
            raise(Exception("User is None"))
        if self.current_user not in self.memory:    
            self.memory[self.current_user] = [fact]
        else:
            self.memory[self.current_user] += [fact]
        self.forget()

    def inquire_name(self) -> str:
        name = input("What is your name? ")
        while type(name) != str:
            name = input("Name has to be str, please try again! ")
        return name

    def get_fact(self) -> str:
        fact = input("Please specify the fact ")
        while type(fact) != str:
            fact = input("Fact has to be str, please try again! ")
        return fact

    def forget(self) -> None:
        forget_threshold = 0.05
        prob = random.random()
        if prob <= forget_threshold:
            #choose a random person
            choice = random.choice(list(self.memory.keys()))
            #forget, or mistake:
            if random.random() >= 0.5:
                #forget
                #choose the item to forget
                index = random.randrange(0, len(self.memory[choice]))
                del self.memory[choice][index]
            else:
                #mistake
                #choose the second person
                choice2 = random.choice(list(self.memory.keys()))
                while choice == choice2:
                    choice2 = random.choice(list(self.memory.keys()))
                #choose the first item
                index1 = random.randrange(0, len(self.memory[choice]))
                index2 = random.randrange(0, len(self.memory[choice2]))

                #swap facts
                self.memory[choice][index1], self.memory[choice2][index2]\
                 = self.memory[choice2][index2], self.memory[choice][index1]

    def step(self):
        choice = input("Hmmm, would you like to hear facts about yourselves, or add a fact, or quit? ")
        while choice not in self.options:
            choice = input(f"Please choose one of the options!: {self.options} ")
        if choice == "Add a fact":
            self.update_mem(self.get_fact())
            while input("Added to memory, another fact? ") == "Yes":
                self.update_mem(self.get_fact())
        elif choice == "Facts about myself":
            if self.current_user not in self.memory:
                print("You haven't told me anything about yourself!")
                return
            else:
                print("Apparently, you:")
                for fact in self.memory[self.current_user]:
                    print(fact)

        elif choice == "Quit":
            print("Okayy, see u!")
            self.current_user = self.inquire_name()
            
#// END_TODO
```

```python
chat_bot = Bot()

while True:
    chat_bot.step()
```
