## X. Multiple Player Horse Betting


- This is an advanced version of the normal Horse Betting game - 
Want to play a betting game but don't want to lose any real money? Create it yourself!
You start with a wallet that contains 100. Each round you get to choose on which horse you want to bet. You also choose the amount you want to bet. 
If the horse you chose is in first place (after shuffling the list) then the amount you put in, will be doubled. Otherwise you lose your money.
The bets are prompted using the input() function. The player will be asked to guess a letter continuously without having to rerun the code every time. If the amount in your wallet is 0 or the player has as input -1, the betting will end immediately.

Now you will have to make this in multiplayer function:
The inputs you take first are the number of players and the names of the players.

**Hint:**

To shuffle a list, you first have to import the random library with ``import random``. Then you can shuffle a list using ``random.shuffle(horses)``

**Notes:**
1. You will create 2 functions:
    - The function ``start_race()``, which takes the horse of choice, amount that has been betted on this horse, and the total amount that is in the wallet as input.
    - The function ``create_players()``, which takes as input parameters the number of players and the amount of money in their start wallet. It should return a list of dictionaries in which all information of the player can be stored.
2. First a list of the horses with numbers in front will be printed.
3. It will then be asked how many players there. After the user has filled this in, the function ``create_players()`` will be called. In the function the name of the player is asked and stored into the dict.
4. It will then ask each of the players to insert the number of the horse they want to choose.
    <pre>"You chose: Affirmed."</pre>
4. Then it will ask the player to insert the amount they want to bet:
    <pre>"You have betted: amount."</pre>
    - If the player does not have that amount in their wallet, it will print <pre> "You do not have that amount in your wallet."<pre>
    - Else it will call the function ```start_race()```.
5. Shuffle the list in the function.
6. When the player betted on the correct horese, it will print for each player:
    <pre>"You have won the bet. Horse has finished first. You have won: (amount the player put in * 2). Your wallet now contains: (wallet + (amount the player put in * 2)) "</pre>
7. When the player betted on the wrong horese, it will print:
    <pre>"You have lost the bet.  Horse has finished first. You have lost: (amount the player put in). Your wallet now contains: (wallet - amount the player put in)!</pre>
8. Then the program ends. The list of users and their wallets with the correct ammount will be printed for each of them.
